```python
wordcloud = WordCloud(
    background_color='Black',    # Background color of the word cloud
    width=800,                   # Width of the image
    height=800,                  # Height of the image
    max_words=100,               # Max number of words to display
    colormap='plasma',         # Color scheme (can be 'viridis', 'plasma','coolwarm' etc.)
    contour_color='black',       # Contour color around the word cloud
    contour_width=1,             # Contour width
    prefer_horizontal=0.8,       # Horizontal word orientation preference
    min_font_size=10,            # Minimum font size for words
    max_font_size=100,           # Maximum font size for words
    font_path=None               # Path to a custom font (optional)
).generate(data)

# Display the generated word cloud
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide the axis
plt.show()
```
